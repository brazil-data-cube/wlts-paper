### Trajectory database

Considering the number of points that need to be extracted, it may take a considerable time to finish the process. If some problem occurs, the program may stop running in this process, preventing the extraction from continuing. In such cases, the extraction must be done from the beginning.

To avoid this problem, in this notebook, each extracted trajectory is saved in a database. Thus, when the program needs to be restarted, the database determines from which point the program will be started. This strategy is applied using the database [TinyDB](https://tinydb.readthedocs.io/en/latest/).

The TinyDB instance is configured below:

```python
from tinydb import TinyDB  # library to use the tinydb

db = TinyDB(os.path.join(output_directory, "harmonized-trajectories_mapbiomas-terraclass_2014.json"), ensure_ascii=False)
```


Additionally, a log file is also created that registers all the extractions that were performed. This log allows the control of what was done and the later visualization of the found errors. In this notebook, we will create the log with the [Loguru](https://github.com/Delgan/loguru) library. The log is configured below:


```python
from loguru import logger  # library to create logs

logger.add(os.path.join(output_directory, "harmonized-trajectories_download_{time}.log"))
```

## Retrieving trajectories
<hr style="border:1px solid #0077b9;">

To begin retrieving the trajectories, we will first load the libraries that we will use:


```python
import folium              # library to create interactive maps
import geopandas as gpd    # library to manipulate vector data (Loaded with the alias `gpd`)

from wlts import WLTS      # library to access the WLTS service
```

### Sample points

Now, we will use the `read_file` function from the [geopandas](https://geopandas.org/) library to load the shapefile with the grid points that we generated for the So Felix do Xingu region.

```python
sample_points = gpd.read_file("analysis/data/raw_data/study-area_sao-felix-do-xingu/sao-felix-do-xingu_utm_sqr_pts1km.shp")
sample_points
```

**Visualizing the points**

Using the [folium](https://python-visualization.github.io/folium/) library, we will now visualize the grid points loaded in an interactive map:

```python
#
# Creating the base map
#
m = folium.Map(
    location = [-6.512413, -52.68641],
    zoom_start = 10,
)

#
# Adding the satellite imagery in the background
#
folium.TileLayer(
      tiles   = "https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
      attr    = "Esri",
      name    = "Esri Satellite",
      overlay = False,
      control = True
).add_to(m)

#
# Adding the grid points in the map
#
for point in sample_points.geometry:
    folium.CircleMarker(
        location=[ point.xy[1][0], point.xy[0][0] ],
        radius=2,
        color="blue",
        fill=False,
    ).add_to(m)

m
```

### Defining the WLTS Collections

To conduct the case study, we will perform trajectory extraction considering two collections:

- `MapBiomas Amaznia v5`;
- `TerraClass Amaznia v2`.

These collections have specific identifiers in WLTS, so we need to query the service to retrieve these identifiers. To do this, we can use the `collection` attribute from the WLTS object available on [wlts.py](https://github.com/brazil-data-cube/rwlts) package, which queries the collections available in a given WLTS service:

```python
#
# Creating an WLTS object
#
service = WLTS(wlts_service_url, access_token = bdc_access_token)

#
# Listing collections from WLTS service
#
service.collections
```

Based on this information, below each collection is linked to its identifier in WLTS. In the code below, a variable with the reference to these collections identifiers is defined:

- `MapBiomas Amaznia v5 (mapbiomas_amazonia-v5)`;
- `TerraClass Amaznia v2 (terraclass_amazonia-v2)`.


```python
collections = "terraclass_amazonia-v2,mapbiomas_amazonia-v5"
```

### Defining the temporal extent

The retrieving of the trajectories will be made considering data between the years `2004` and `2014`:

```python
start_date = 2014
end_date   = 2014
```

### Retrieving the LULC Harmonized trajectories

Now, we will retrieve trajectories for each of the points that were loaded previously. The points will be harmonized using the `Simplified-legend-TM-1` mapping.

> The `Simplified-legend-TM-1` is the same that the simplified mapping presented by Zioti *et al.* (2021).


```python
#
# Defining the last TinyDB registered ID, if available
#
last_row = db.get(doc_id=len(db))
last_row

#
# Trajectories extraction
#
for index, row in sample_points.iterrows():
    
    logger.info(f'row({index})')

    #         
    #  Check if this location has already been consulted. If it is a new location, the next step is performed
    # 
    if last_row and index <= last_row["id"]:
        continue
   
    #
    #  Retrieve the LULC trajectory from the WLTS service    
    # 
    _res = service.tj(latitude      = row.geometry.y,
                      longitude     = row.geometry.x,
                      collections   = collections,
                      start_date    = start_date,
                      end_date      = end_date,
                      target_system = "Simplified-legend-TM-1")

    # 
    # Save the result on TinyDB
    # 
    db.insert({"id": index, **_res})
```
